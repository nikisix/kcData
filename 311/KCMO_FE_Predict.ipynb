```python
# read in data
loc = '/Users/pferrante/Documents/Data/ODKC/'
data = []

for i in os.listdir(loc):
    if 'monthly' in i:
        print i
        pth = loc+i
        data.append(pd.read_csv(pth, compression='gzip'))

df = pd.concat(data)
```

```python
# filter for 4 years
df=df[(df['year']>=2011)]
df=df[(df['month']!=1)]
print df.shape
```

```python
# sum concurrent water-related call records
cols=[c for c in df.columns.values.tolist() if (('concurrent_' in c and 'Maintenance' in c) or ('concurrent_' in c and 'Engineering' in c))]
df.insert(0, 'call_count2', df[cols].apply(np.sum, axis=1))

# binarize dependent variable
df.insert(0, 'call_count', df['call_count2'].map(lambda x: 1 if x > 0 else 0))
cols.append('call_count2')
print df.shape
```

```python
# check dependent class
print df['call_count'].value_counts()
sns.factorplot("call_count", data=df, palette="PuBuGn_d");
```

```python
CONTINUOUS_FIELDS = []
FACTOR_FIELDS = []
IMPUTE_FIELDS = []
MISSING_FIELDS = []
DROP_FIELDS = []
loc = {}

# classify fields
for i,col in enumerate(df.columns):
    dataType = df[col].dtype
    missingCount = float(pd.isnull(df[col]).sum())
    missingPct = missingCount / float(len(df[col]))
    uniqueCount = df[col].nunique()
    loc[col] = i
    if (missingPct > 0.0 and missingPct < 0.5):
        IMPUTE_FIELDS.append(col)    
    elif missingPct >= 0.5:
        MISSING_FIELDS.append(col)
    elif (dataType == object and uniqueCount == 1):
        DROP_FIELDS.append(col)
    elif (dataType == 'float64' or dataType == 'int64'):
        CONTINUOUS_FIELDS.append(col)
    else:
        FACTOR_FIELDS.append(col)

# create data source specific lists
other_cols = ['onewaycode', 'snow_route', 'speed_limit', 'yearconstr']
w1 = loc['Multiday_precipitation_total__tenths_of_mm__']
w2 = loc['Peak_gust_time__HHMM___lag']
weather_cols = df.columns[w1:w2+1].values.tolist()
c1 = loc['Area_Total']
c2 = loc['Vacant_housing_units__Other_vacant']
census_cols = df.columns[c1:c2+1].values.tolist()
```

```python
# print fields
tempStr = ''
for m in MISSING_FIELDS:
    tempStr += m + ', '
print 'MISSING_FIELDS: ' + tempStr
tempStr = ''
for i in IMPUTE_FIELDS:
    tempStr += i + ', '
print 'IMPUTE_FIELDS: ' + tempStr
tempStr = ''
for d in DROP_FIELDS:
    tempStr += d + ', '
print 'DROP_FIELDS: ' + tempStr
```

```python
for col in IMPUTE_FIELDS:
    # impute based on monthly mean for weather
    if col in weather_cols:
        df[col] = df.groupby(['year','month'])[col].transform(lambda x: x.fillna(x.mean()))
    # impute based on monthly median for other
    if col in other_cols:
        df[col] = df.groupby(['year','month'])[col].transform(lambda x: x.fillna(x.median()))

for col in MISSING_FIELDS:
    # create a dummy variable for highly missing
    if col in weather_cols:
        newCol = col + '__na_check'
        df[newCol] = df[col].apply(lambda x: 1 if x == np.nan else 0)
        del df[col]
```

```python
CONTINUOUS_FIELDS = []
FACTOR_FIELDS = []
IMPUTE_FIELDS = []
MISSING_FIELDS = []
DROP_FIELDS = []
loc = {}

# reclassify fields
for i,col in enumerate(df.columns):
    dataType = df[col].dtype
    missingCount = float(pd.isnull(df[col]).sum())
    missingPct = missingCount / float(len(df[col]))
    uniqueCount = df[col].nunique()
    loc[col] = i
    if (missingPct > 0.0 and missingPct < 0.5):
        IMPUTE_FIELDS.append(col)    
    elif missingPct >= 0.5:
        MISSING_FIELDS.append(col)
    elif (dataType == object and uniqueCount == 1):
        DROP_FIELDS.append(col)
    elif (dataType == 'float64' or dataType == 'int64'):
        CONTINUOUS_FIELDS.append(col)
    else:
        FACTOR_FIELDS.append(col)
```

```python
# create a dummy variable for remaining
for col in IMPUTE_FIELDS:    
    if col in weather_cols:
        newCol = col + '__na_check'
        df[newCol] = df[col].apply(lambda x: 1 if x == np.nan else 0)
        del df[col]
```

```python
# generate roc curve
lg1_fpr, lg1_tpr, lg1_thresholds = roc_curve(target_test, lg1_probs[:, 1])
lg1_roc_auc = auc(lg1_fpr, lg1_tpr)
lg2_fpr, lg2_tpr, lg2_thresholds = roc_curve(target_test, lg2_probs[:, 1])
lg2_roc_auc = auc(lg2_fpr, lg2_tpr)
#sgd_fpr, sgd_tpr, sgd_thresholds = roc_curve(target_test, sgd_probs[:, 1])
#sgd_roc_auc = auc(sgd_fpr, sgd_tpr)

plt.plot(lg1_fpr, lg1_tpr, label='Logistic Regression #1 (area = %0.2f)' % lg1_roc_auc)
plt.plot(lg2_fpr, lg2_tpr, label='Logistic Regression #2 (area = %0.2f)' % lg2_roc_auc)
#plt.plot(sgd_fpr, sgd_tpr, label='Stochastic Gradient Descent Classifier (area = %0.2f)' % sgd_roc_auc)

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
```
